Apply corelate2d

```python
from matplotlib import pyplot as plt
from skimage.color import rgb2gray
from scipy.signal import convolve2d, correlate2d
import numpy as np

# load image and convert to grayscale image
im = plt.imread('london.png')
im = rgb2gray(im)

# create kernel
n = 11
kernel = np.ones((n,n))/(n*n)
# print(kernel)

# apply correlation
out = correlate2d(im, kernel, mode='same')

plt.subplot(121);plt.imshow(im, cmap=plt.get_cmap('gray'))
plt.title('Original Imgae'), plt.xticks([]), plt.yticks([])
# show image

plt.subplot(122);plt.imshow(out, cmap=plt.get_cmap('gray'))
plt.title('After using correlate2d'), plt.xticks([]), plt.yticks([])
plt.show()

# print size of input and output
print(im.shape)
print(out.shape)
```
